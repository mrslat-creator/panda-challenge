```python
# Dependencies and Setup
import pandas as pd
import numpy as py

# File to Load
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data_df = pd.DataFrame(purchase_data)
purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
buy_players = purchase_data_df["SN"].count()
buy_players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Find unique items, total purchases and average prices of items

unique_items = len(purchase_data_df["Item ID"])
unique_items

unique_items2 = len(purchase_data_df["Item Name"])
unique_items2

total_items_purchased = purchase_data_df["Purchase ID"].count()
total_items_purchased

total_money_made = purchase_data_df["Price"].sum()
total_money_made

average_amount1 = purchase_data_df["Price"].mean()
average_amount1

average_amount2 = total_money_made/total_items_purchased
average_amount2

# Display the summary data frame
purchase_analysis_df = pd.DataFrame([{"Number of Unique Items": unique_items, "Average Amount": average_amount1,
                                      "Number of Purchases": total_items_purchased, "Total Money Made": total_money_made}])
purchase_analysis_df["Average Amount"] = purchase_analysis_df["Average Amount"].map("${:,.2f}".format)
purchase_analysis_df["Total Money Made"] = purchase_analysis_df["Total Money Made"].map("${:,.2f}".format)
purchase_analysis_df

# Reordering the column order for summary dataframe
org_purchase_analysis_df = purchase_analysis_df[["Number of Unique Items","Average Amount","Number of Purchases", "Total Money Made" ]]
org_purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Purchase Data by Gender
gender_stat = purchase_data.groupby("Gender")

#Total count of screen names by gender
gender_count_total = gender_stat.nunique()["SN"]

#Total count by gender and divided by total players
percentage_of_total_players = gender_count_total/buy_players*100

#Create dataframe with combined values for demographics
gender_demo = pd.DataFrame({"Percentage of Total Players": percentage_of_total_players, "Total Count": gender_count_total})

#Format the values sorted by total count in decending order, and percentage with two decimal places
gender_demo.sort_values(["Total Count"], ascending = False).style.format({"Percentage of Players":"{:.2f}"})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Using GroupBy in order to separate the data into fields according to "Gender" values
gender_grouped_purchased_data_df = purchase_data_df.groupby(["Gender"])

# Visual of gender grouped purchase data
gender_grouped_purchased_data_df["Purchase ID"].count().head(10)

#Get total purchase value grouped by gender
total_purchase_value = gender_grouped_purchased_data_df["Price"].sum()
total_purchase_value.head()
dlr_total_purchase_value = total_purchase_value.map("${:,.2f}".format)
dlr_total_purchase_value.head()

#Average Price by gender
avg_purchase_price = gender_grouped_purchased_data_df["Price"].mean()
avg_purchase_price.head()
dlr_avg_purchase_price = avg_purchase_price.map("${:,.2f}".format)
dlr_avg_purchase_price.head()

# Organize summary gender data, get all columns to organized Data Frame, add needed columns to it
original_gender_purchased_data_df = pd.DataFrame(gender_grouped_purchased_data_df["Purchase ID"].count())
original_gender_purchased_data_df["Average Purchase Price"] = dlr_avg_purchase_price  
original_gender_purchased_data_df["Total Purchase Value"] = dlr_total_purchase_value 
original_gender_purchased_data_df

#Summary purchasing analysis Dataframe grouped by gender, rename "Purchase"
summary_gender_purchased_data_df = original_gender_purchased_data_df.rename(columns={"Purchase ID":"Purchase Count"})
summary_gender_purchased_data_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Purchase Data will be held
# Bins are 0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]

# Create the names for the four bins
grp_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

group_by_age_purchase_data_df = purchase_data_df
group_by_age_purchase_data_df["Age Summary"] = pd.cut(group_by_age_purchase_data_df["Age"], age_bins, labels=grp_names)
group_by_age_purchase_data_df

#Create a group based off of the bins and saving it
group_by_age_purchase_data_df = group_by_age_purchase_data_df.groupby("Age Summary")
group_by_age_purchase_data_df.count()

#Summary of purchase data in a new dataframe
summ_by_age_df = pd.DataFrame(group_by_age_purchase_data_df.count())
summ_by_age_df

# Calculations performed on "Purchase ID" column of Summary dataframe
summ_by_age_df["Purchase ID"] = (summ_by_age_df["Purchase ID"]/buy_players)*100
summ_by_age_df

# Format the percentages of summary by age dataframe
summ_by_age_df["Purchase ID"] = summ_by_age_df["Purchase ID"].map("{:,.2f}%".format)
summ_by_age_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Sorted data analysis by age new dataframe
analysis_by_age_df = pd.DataFrame(group_by_age_purchase_data_df["Purchase ID"].count())
analysis_by_age_df

# Total purchase value sorted by age
total_purchase_value_age = group_by_age_purchase_data_df["Price"].sum()
total_purchase_value_age
dlr_total_purchase_value_age = total_purchase_value_age.map("${:,.2f}".format)
dlr_total_purchase_value_age

# Average purchase price by age
avg_purchase_price_age =group_by_age_purchase_data_df["Price"].mean()
avg_purchase_price_age
dlr_avg_purchase_price_age = avg_purchase_price_age.map("${:,.2f}".format)
dlr_avg_purchase_price_age

# Get normal totals by age, total purchase value divided by
norm_totals_age = total_purchase_value_age/group_by_age_purchase_data_df["Purchase ID"].count()
dlr_norm_totals_age = norm_totals_age.map("${:,.2f}".format)
dlr_norm_totals_age

# Organize summary gender data, get all columns to organized Data Frame, add needed columns to it
analysis_by_age_df["Average Purchase Price"] = dlr_avg_purchase_price_age  
analysis_by_age_df["Total Purchase Value"] = dlr_total_purchase_value_age 
analysis_by_age_df["Normalized Totals"] = dlr_norm_totals_age 
analysis_by_age_df

# Summary of purchasing analysis dataframe grouped by age, rename "Purchase"
summ_age_purchased_data_df = analysis_by_age_df.rename(columns={"Purchase ID":"Purchase Count"})
summ_age_purchased_data_df
```
